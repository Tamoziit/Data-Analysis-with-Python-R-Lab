```python
import math

class RegTooSmall(Exception):
    pass

class RegTooHigh(Exception):
    pass

def find_n(n):
    res = math.ceil(math.log2(n))
    return res

def check_range(n):
    if n > 255:
        raise RegTooHigh("Too many reg.")
    elif n < 1:
        raise RegTooSmall("Too less reg.")
    else:
        return find_n(n)

while True:
    try:
        n = int(input("Enter no. of reg.: "))
        res = check_range(n)
        print(res)
        break
    except ValueError as e:
        print(e)
    except RegTooSmall as e:
        print(e)
    except RegTooHigh as e:
        print(e)
    except Exception as e:
        print(e)
```

```python
import re

s = input("Enter a number: ")

pattern = r'^[+-]?\d+(\.\d+)?$'
if re.match(pattern, s):
    num = float(s)
    print(num ** 2.0)
else:
    print("Invalid")
```

```python
import math

class Point:
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y
    
    def dist(self):
        dist = math.sqrt(self.x ** 2 + self.y ** 2)
        print(dist)
    
    def display(self):
        print(f'({self.x}, {self.y})')

ob = Point(9, 8)
ob.display()
ob.dist()
```
