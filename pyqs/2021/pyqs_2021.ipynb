```python
x = int(input("Enter x-coord: "))
y = int(input("Enter y-coord: "))
print(f"({x}, {y})")

if x > 0 and y > 0:
    print("1st Quad.")
elif x < 0 and y > 0:
    print("2nd Quad.")
elif x > 0 and y < 0:
    print("4th Quad.")
elif x < 0 and y < 0:
    print("3rd Quad.")
elif x == 0 and y == 0:
    print("Origin")
elif x == 0:
    print("Y-axis")
elif y == 0:
    print("X-axis")
```

```python
df = pd.DataFrame({
    "M0": [8, np.nan, np.nan, 7],
    "M1": [2, 2, np.nan, 3],
    "M2": [np.nan, 3, np.nan, 7]
})
print(df)

df2 = df.fillna(0)
print(df2)

df3 = df.ffill()
print(df3)

df4 = df.fillna(df.mean(numeric_only=True))
print(df4)
```

```python
data = {
    "City": ["Delhi", "Kolkata", "Mumbai"],
    "2020": [3.029, 1.485, 2.0411],
    "2019": [2.94, 1.4755, 2.0185]
}

df = pd.DataFrame(data)
print(df)

df['2018'] = [2.85, 1.468, 1.998]
print(df)

df["Average"] = df[["2020", "2019", "2018"]].mean(axis=1)
print(df)
```
