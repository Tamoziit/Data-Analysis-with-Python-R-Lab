```python
# 3a
def is_abecedarian(s):
    # If length is 0 or 1 â†’ automatically abecedarian
    if len(s) <= 1:
        return True
    
    # Check if first two characters are in correct order
    if s[0] > s[1]:
        return False
    
    # Recur on the remaining substring
    return is_abecedarian(s[1:])


lst = ['accddmop', 'addccnm']
res = is_abecedarian(lst)
print(lst)
print(res)
```

```python
# 3d - Binary Search
def bin_search(arr, ele):
    l = 0
    h = len(arr) - 1
    
    while (l < h):
        mid = (l + h) // 2
        if (arr[mid] == ele):
            print("Ele. found")
            return True
        elif (arr[mid] > ele):
            h = mid - 1
        elif (arr[mid] < ele):
            l = mid + 1
    
    print("Ele. not found")
    return False

arr = eval(input("Enter a list: "))
ele = int(input("Enter no. to search: "))

print(arr, ele)
bin_search(arr, ele)
```

```python
# 4a
def fun(n):
    if n == 0:
        return "0"
    elif n == 1:
        return "1"
    else:
        return fun(n//2) + fun(n%2)

res = fun(7)
print(res)
print(type(res))
```

```python
# 4b
def pizza_pieces(n):
    if n==0:
        return 1
    else:
        return pizza_pieces (n-1) + n

print(pizza_pieces(9))

def factorial(n):
    if n==0:
        return 1
    else:
        return factorial (n-1) * n

print(factorial(5))
```

```python
# 7
data = {
    "Name": ["Abhijit", "Lata", "Asha", "Preetam"],
    "Physics": [85, 80, 89, 88],
    "Chemistry": [90, 92, 80, 93],
    "Mathematics": [92, 99, 82, 95]
}

df = pd.DataFrame(data)
print(df)

df["Biology"] = [88, 82, 80, 79]
print(df)

df["Total"] = df[["Physics", "Chemistry", "Mathematics", "Biology"]].sum(axis=1)
print(df)

subjects = ["Physics", "Chemistry", "Mathematics", "Biology"]

totals = [df[sub].sum() for sub in subjects]
means = [df[sub].mean() for sub in subjects]
medians = [df[sub].median() for sub in subjects]
sds = [df[sub].std() for sub in subjects]


df2 = pd.DataFrame({
    "Subject": subjects,
    "Total": totals,
    "Mean": means,
    "Median": medians,
    "SD": sds
})
print(df2)
```
