```python
# 3a
def is_abecedarian(s):
    # If length is 0 or 1 â†’ automatically abecedarian
    if len(s) <= 1:
        return True
    
    # Check if first two characters are in correct order
    if s[0] > s[1]:
        return False
    
    # Recur on the remaining substring
    return is_abecedarian(s[1:])


lst = ['accddmop', 'addccnm']
res = is_abecedarian(lst)
print(lst)
print(res)
```

```python
# 3c
def swap(lst):
    first, *middle, last = lst
    return [last, *middle, first]

lst = [10, 20, 30, 40]
res = swap(lst)
print(lst, res)
```

```python
# 3d - Binary Search
def bin_search(arr, ele):
    l = 0
    h = len(arr) - 1
    
    while (l < h):
        mid = (l + h) // 2
        if (arr[mid] == ele):
            print("Ele. found")
            return True
        elif (arr[mid] > ele):
            h = mid - 1
        elif (arr[mid] < ele):
            l = mid + 1
    
    print("Ele. not found")
    return False

arr = eval(input("Enter a list: "))
ele = int(input("Enter no. to search: "))

print(arr, ele)
bin_search(arr, ele)
```
