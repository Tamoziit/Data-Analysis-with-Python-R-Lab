## Misc

```python
nums = [2, 4, 5, 6, 11, 16]

res = [x for x in nums if x % nums[0] != 0 or x == nums[0]]
res
```

```python
x = [1, 2, 3, 4, 5]

res = list(map(lambda x: x ** 3.0, filter(lambda item: item % 2 != 0, x)))
res
```

```python
nums = list(range(1, 19))
print(nums)

sq_lst = list(map(lambda x: x ** 2.0, nums))
print(sq_lst)

div = list(filter(lambda x: x % 5 == 0, nums))
print(div)

from functools import reduce
sum = reduce(lambda a, b: a + b, nums)
print(sum)
```
