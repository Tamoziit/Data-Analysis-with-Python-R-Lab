```python
# Re-indexing
a = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
print(a)

b = a.reindex(['a', 'd', 'b', 'c'])
print(b)

df = pd.DataFrame({
    "Marks": [33, 34, 35]
}, index=["Ram", "Sohan", "Krishn"])
print(df)

df2 = df.reindex(["Ram", "Sohan", "Krishn", "Alex"])
print(df2)
```

```python
arr = np.array(["hello", "WORLD", "pYtHoN"])

print("Original:", arr)

print(np.char.capitalize(arr))
print(np.char.lower(arr))
print(np.char.upper(arr))
print(np.char.swapcase(arr))
print(np.char.title(arr))
```

```python
import random

class NumberSmall(Exception):
    pass

try:
    n = random.random()
    print(n)
    if n < 0.5:
        raise NumberSmall("No. too small!")
except NumberSmall as e:
    print(e)
```

```python
def divide(m, n):
    if m > n or n <= 0 or m <= 0:
        raise ValueError("Enter valid nos.")
    return [i for i in range(1, n+1) if i % m == 0]

m = int(input("Enter m: "))
n = int(input("Enter n: "))
try:
    res = divide(m, n)
    print(res)
except ValueError as e:
    print(e)
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 4, 5, 6, 7]

res = [item for item in x if item in y]
print(res)
```

```python
res2 = list(filter(lambda item: item in y, x))
print(res2)
```

```python
text = "Hello World"

char_count = {}
for ch in text:
    if ch.isalpha():
        char_count[ch] = char_count.get(ch, 0) + 1

for ch in sorted(char_count):
    print(f'{ch}: {char_count[ch]}')
```
